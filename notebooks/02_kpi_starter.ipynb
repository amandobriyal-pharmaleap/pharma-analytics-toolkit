# KPI Starter Notebook

Compute baseline KPIs for pharma sales performance.


```python
import pandas as pd

df = pd.read_csv('../data/sample_sales.csv')
kpis = {
    'total_sales': float(df['sales'].sum()),
    'avg_sales_per_brand': float(df.groupby('brand')['sales'].sum().mean()),
    'unique_doctors': int(df['doctor_id'].nunique()),
}
kpis
```
